```python
#import pandas
import pandas as pd
```

```python
#import in the csv with all restaurants
restaurants_csv = "Resources/all-restaurants.csv"
```

```python
#read in the csv as a dataframe
restaurant_df = pd.read_csv(restaurants_csv)
restaurant_df.head()
```

```python
#read in the columns
restaurant_df.columns
```

```python
#create a new data frame and drop the unnamed column
new_rest_df = restaurant_df[["Name", "Rating", "Review Count", "Street Address",
                                    "Zip Code", "Url", "Latitude", "Longitude"]]
```

```python
#rename the columns to no longer have spaces and so they match our database
rename = new_rest_df.rename({'Zip Code': 'Postcode', 'Review Count': 'Num_Reviews','Street Address' : 'street_address'}, axis=1)  # new method
```

```python
rename.head()
```

```python
#create a new table with just location information
restaurant_location = rename[["Name","street_address","Postcode","Latitude","Longitude"]]
restaurant_location.head()
```

```python
#create a new table with just additional info
restaurant_info = rename[["Name","street_address","Rating","Num_Reviews","Url"]]
restaurant_info.head()
```

```python
#save both of these dataframes as csvs
restaurant_location.to_csv("../final-csvs/restaurant-location.csv")
restaurant_info.to_csv("../final-csvs/restaurant-info.csv")
```
